```
import tensorflow as tf
from tensorflow import keras

# Helper libraries
import numpy as np
np.random.seed(42)
import matplotlib.pyplot as plt
from tensorflow.keras import datasets, layers, models
from keras.models import Sequential
print(tf.__version__)
```

**So for MNIST, make sure your images have shape (28, 28, 1), not (784,)!**

```
from keras.datasets import mnist
(train_images, train_labels), (test_images, test_labels) = mnist.load_data()
print(train_labels[0])
plt.imshow(train_images[0])

#data = tf.data.Dataset.from_tensor_slices(
    #(train_images.reshape([-1, 28, 28, 1]).astype(np.float32) / 255, train_labels.astype(np.int32)))
#data = data.shuffle(buffer_size=60000).batch(128).repeat()

#test_data = tf.data.Dataset.from_tensor_slices(
    #(test_images.reshape([-1, 28, 28, 1]).astype(np.float32) / 255, test_labels.astype(np.int32))).batch(10000)

# First we need to rescale our test and train image data
#train_images = train_images/ 255.0
#test_images = test_images/ 255.0
```

```

#Since for CNN we need to pass the input of number of images we have and what are the pixels , so we need to convert the data into 4 D
#train_images = train_images.reshape(len(train_images),28,28,1)
#test_images = test_images.reshape(len(test_images),28,28,1)
train_images = train_images.reshape([-1, 28, 28, 1]).astype(np.float32) / 255
train_labels = train_labels.reshape((-1,)).astype(np.int32)

test_images = test_images.reshape([-1, 28, 28, 1]).astype(np.float32) / 255
test_labels = test_labels.reshape((-1,)).astype(np.int32)


train_data = tf.data.Dataset.from_tensor_slices((train_images, train_labels))
train_data = train_data.shuffle(buffer_size=60000).batch(1000)

test_data = tf.data.Dataset.from_tensor_slices((test_images, test_labels))
test_data = test_data.batch(1000)
```

```
#from keras.layers import Flatten,Conv2D,Dense
train_steps = 2500
model = models.Sequential()

model.add(layers.Conv2D(32,kernel_size=3,activation='relu',input_shape=(28,28,1),padding="SAME"))
act_map = model.get_output_at(0)
model.add(layers.MaxPooling2D((2, 2)))
model.add(layers.Conv2D(64, kernel_size=3, activation='relu',padding="SAME"))
model.add(layers.MaxPooling2D((2, 2)))
model.add(layers.Conv2D(64, kernel_size=3, activation='relu',padding="SAME"))
# Now we will add the output to the dense layer to perform the classification but the dense layer takes the input 1d but we are having ouput 3D ( 11,11,32) so we will flatten
model.add(layers.Flatten())
model.add(layers.Dense(64,activation='relu'))
model.add(layers.Dense(10))


loss_func= tf.losses.SparseCategoricalCrossentropy(from_logits=True)
#model.compile(optimizer='adam', loss=loss_func, metrics=['accuracy'])
```

```
model.summary()
```

```
optimizer1 = tf.optimizers.Adam()
```

```
varis = model.trainable_variables

loss_list = []
for step, (img_batch, lbl_batch) in enumerate(train_data):
    if step > train_steps:
        break

    with tf.GradientTape() as tape:
        logits = model(img_batch)
        xent = loss_func(lbl_batch, logits)
        loss_list.append(xent.numpy)

    grads = tape.gradient(xent, varis)

    optimizer1.apply_gradients(zip(grads, varis))
    
    if not step % 100:
        preds = tf.argmax(logits, axis=1, output_type=tf.int32)
        acc = tf.reduce_mean(tf.cast(tf.equal(preds, lbl_batch),
                             tf.float32))
        print("Loss: {} Accuracy: {}".format(xent, acc))
        

    
```

```
plt.plot(loss_list)
plt.xlabel('Training Steps')
plt.ylabel('Loss')
```

```
acc_list = []
for step, (img_batch, lbl_batch) in enumerate(test_data):
  logits = model(img_batch)
  preds = tf.argmax(logits, axis=1, output_type=tf.int32)
  acc = tf.reduce_mean(tf.cast(tf.equal(preds, lbl_batch), tf.float32))
  acc_list.append(acc.numpy())
  print(acc)

print("Average test accuracy: ", np.mean(acc_list))
```

```

plt.plot(acc_list)
plt.xlabel('Training Steps')
plt.ylabel('Accuracy')
```
